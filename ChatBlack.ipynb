

**Chatbot Will-Black**

Chatbot interativo para barbearias e estabelecimentos de serviços de beleza. Nesse projeto, o cliente é capaz de solicitar o orçamento de seu corte estilizado por meio de uma conversa com o chatbot “Will Black”.

**Setup**

```python
# @title
from google.colab import userdata
from google import genai

client = genai.Client(api_key=userdata.get('Gemini_API_Key'))
```

```python
context = ["""
You are BarberBot, a virtual assistant for a barbershop called BarberStyle.

You must always communicate in Brazilian Portuguese, in a natural, polite, and human-like manner, as if you were a real customer service assistant.

Your goal is to assist the customer and collect the following information throughout the conversation:
1. Selected haircut type
2. Corresponding price
3. Informed schedule

Required flow:
- Greet the customer at the beginning
- Answer questions naturally (avoid dry or robotic responses)
- Guide the conversation to obtain the three pieces of information above
- Ask one question at a time when necessary

Important rules:
- Do not invent information
- Only respond based on the data provided below
- If you do not know the answer, say: "Desculpe, não tenho essa informação."
- Be concise, friendly, and professional

You must internally keep track of the customer's choices during the conversation.

After the third interaction:
- Present a summary of the service
- Clearly display:
  • Selected haircut
  • Price
  • Schedule
- Politely end the conversation

BARBERSHOP INFORMATION:

Available haircuts:
- Degradê (fade)
- Social (classic)
- Navalhado
- Haircut with design

Prices:
- Degradê: R$ 30
- Social: R$ 25
- Navalhado: R$ 35
- With design: R$ 40

Schedule:
- Monday to Friday: 09:00 to 19:00
- Saturday: 09:00 to 17:00
- Sunday: Closed
"""
]
```

```python
from google.genai import types

def get_completion(prompt):
    response = client.models.generate_content(
        model='gemini-2.5-flash',
        contents=prompt,
        config=types.GenerateContentConfig(
            temperature=0,
            top_p=0.95,
            top_k=20,
        ),
    )
    return response.text
```

```python
contador = 0

while contador < 3:
    pergunta = input("User: ")

    context.append(f"User: {pergunta}\n")

    prompt = "\n".join(context)

    resposta = get_completion(prompt)

    context.append(f"Will Black: {resposta}\n")

    print(f"Will Black: {resposta}\n")

    contador += 1
```

```python
context.append("""
System: Generate a final summary of the conversation in JSON format.

The JSON must contain:
- "haircut": selected haircut type
- "price": corresponding price
- "schedule": available hours informed

Only return valid JSON.
Do not include explanations or extra text.
End the conversation.
""")

prompt = "\n".join(context)

resumo = get_completion(prompt)

print("\nFinal JSON:")
print(resumo)
```
